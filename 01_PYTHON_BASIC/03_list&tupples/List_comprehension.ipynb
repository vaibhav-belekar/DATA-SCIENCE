# list comprehension
List Comprehension is a concise and expressive syntactic construct in Python used to create new lists by iterating over an iterable (such as a list, tuple, string, or range) and optionally applying a transformation and/or filtering condition to each element, all within a single line of code.

It provides an alternative to traditional looping constructs (like for loops with append()), enabling more readable, compact, and often more efficient list creatio


```python
lst1=[]
def lst_square(lst):
    for i in lst:
        lst1.append(i**2)
    return lst1
```

```python
lst=[1,2,3,4,5,6,7,8,9]
lst_square(lst)
```

```python
lst=[1,2,3,4,5,6,7,8,9]
```

```python
[i**2 for i in lst] # this is list comprehension
```

```python
[i**2 for i in lst if i%2==0] # this is list comprehension
```
