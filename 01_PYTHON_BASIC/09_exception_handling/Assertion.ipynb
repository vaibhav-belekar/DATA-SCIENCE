# Python assertionn
The assert statement in Python is used for debugging purposes. It checks whether a given condition is true, and if the condition is false, it raises an AssertionError

```python
num=10
assert num>10
```

```python
try:
    num=int(input("enter even number"))
    assert num%2==0
    print("the number is even")

except AssertionError:
    print(" provide even number")        
 
```
