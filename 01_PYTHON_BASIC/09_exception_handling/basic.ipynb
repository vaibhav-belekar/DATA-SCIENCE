```python
try:
    a=b
    #code block where exception can occur
   
except Exception as e:
    print(e)
```

```python
try:
    a=int(input("eneter a 1st number"))
    b=int(input("eneter a 2nd number"))
    c=a+b
    d=a*b
    e=a/b
    print(c)
    print(d)
    print(e)
    
except ZeroDivisionError as ex3:
    print("the user is trying to divide a number by zero which is not possible")
except NameError as ex1:
    print ("the user is trying to access a variable which is not defined")
except TypeError as ex2:
    print("the user is trying to perform an operation on incompatible data types")
except Exception as e:
    print(e)
```

```python
#try else block
try:
    a=int(input("eneter a 1st number"))
    b=int(input("eneter a 2nd number"))
    c=a+b
    d=a*b
    e=a/b
    print(c)
    print(d)
    print(e)
    
except ZeroDivisionError as ex3:
    print("the user is trying to divide a number by zero which is not possible")
except NameError as ex1:
    print ("the user is trying to access a variable which is not defined")
except TypeError as ex2:
    print("the user is trying to perform an operation on incompatible data types")
except Exception as e:
    print(e)
else:
    print("the code block executed successfully without any exception")
    print("the value of c is",c)
    print(d)
    print(e)
```

```python
#try else block
try:
    a=int(input("eneter a 1st number"))
    b=int(input("eneter a 2nd number"))
    c=a+b
    d=a*b
    e=a/b
    print(c)
    print(d)
    print(e)
    
except ZeroDivisionError as ex3:
    print("the user is trying to divide a number by zero which is not possible")
except NameError as ex1:
    print ("the user is trying to access a variable which is not defined")
except TypeError as ex2:
    print("the user is trying to perform an operation on incompatible data types")
except Exception as e:
    print(e)
else:
    print("the code block executed successfully without any exception")
    print("the value of c is",c)
    print("the value of d is",d)
    print("the value of e is",e)
finally:
    print("the try except block has been executed") 
```

## - custom exception

```python
class Error1(Exception):
    pass
class dobException(Error1):
    pass
```

```python
year=int(input("enter your year of birth"))
age=2024-year
try: 
    
    if age<=30 and age>=20:
        print("you are eligible")
    else:
        raise dobException
except dobException:
        print("you are not eligible")
```
