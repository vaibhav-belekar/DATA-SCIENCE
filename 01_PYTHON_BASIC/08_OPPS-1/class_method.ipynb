# class method and class variable 

```python
class Car:
    base_price=1000 #class variable
    def __init__ (self,window,doors,power):
        self.window=window
        self.doors=doors
        self.power=power
    def what_is_base_price(self):
        print("The base price is {}".format(self.base_price)) 
    @classmethod
    def revised_base_price(cls,inflation):
        cls.base_price=cls.base_price+cls.base_price*inflation
    
    
```

```python
car1=Car(4,5,2000)
```

```python
car1.what_is_base_price()
```

```python
# now we have update inflation rate
car1.revised_base_price(0.10)
```

```python
car1.revised_base_price
```
