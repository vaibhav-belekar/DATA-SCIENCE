# class method and class variable 

```python
class Car:
    base_price=1000 #class variable
    def __init__ (self,window,doors,power):
        self.window=window
        self.doors=doors
        self.power=power
    def what_is_base_price(self):
        print("The base price is {}".format(self.base_price)) 
    @classmethod
    def revised_base_price(cls,inflation):
        cls.base_price=cls.base_price+cls.base_price*inflation
    
    
```

```python
car1=Car(4,5,2000)
```

```python
car1.what_is_base_price()
```

```python
# now we have update inflation rate
car1.revised_base_price(0.10)
```

```python
car1.revised_base_price
```

```python
car1.base_price
```

# static method
“Static methods are used to logically group related functionality inside a class, improving code organization and readability.
- A static method is a method that:

Does not depend on instance (object) data
Does not depend on class data
Works like a normal function, but is placed inside a class


```python
import datetime
now=datetime.datetime.now()
```

```python
now.year
```

```python
# static method
class Car:
    base_price=1000 #class variable
    def __init__ (self,window,doors,power):
        self.window=window
        self.doors=doors
        self.power=power
    def what_is_base_price(self):
        print("The base price is {}".format(self.base_price)) 
        
    @classmethod
    def revised_base_price(cls,inflation):
        cls.base_price=cls.base_price+cls.base_price*inflation
    
    @staticmethod
    def cheaked_year():  
        if now.year==2025:
            return True
        else:
            return False  
```

```python
car1=Car(4,5,5000)
```

```python
car1.cheaked_year()
```

```python
if(car1.cheaked_year()):
    pass
else:
    car1.revised_base_price(0.10)
```
