# python OOPS -Inheritance

```python
#all the classes vasriables are public 
#car blueprint
class Car:
    def __init__(self,window,door,engine):
        self.window=window
        self.door=door
        self.engine=engine
    def drive(self):
        print("the person is driving the car")
```

```python
car=Car(4,5,'petrol')
print(car.window)
```

```python
Car
```

```python
car.drive()
```

```python
class audi(Car):
    def __init__(self, window, door, engine, model):
        super().__init__(window, door, engine)
        self.model = model
    def self_driving(self):
        print("the car is self driving")
```

```python
audi=audi(4,5,'petrol','audi a8')
```

```python
audi.drive()
```

```python
audi.self_driving()
```
