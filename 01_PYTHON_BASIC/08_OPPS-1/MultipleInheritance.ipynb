# multiple Inheritance

```python
 class A:
    def method1(self):
        print("class a method")
```

```python
class B(A):
    def method1(self):
        print("Class B method")
```

```python
class C(A):
    def method1(self):
        print("class c method")
```

```python
class D(B,C):
    def method1(self):
        print("class D method")
        C.method1(self)
        B.method1(self)
        A.method1(self)
        
```

```python
d=D()
```

```python
d.method1()
```

```python
B.method1(d)
```
