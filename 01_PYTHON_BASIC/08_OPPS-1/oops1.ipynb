# oops in python 

```python
class Car:
    def __init__(self,window,door,engine):
        self.window=window
        self.door=door
        self.engine=engine
    def self_driving(self):
        return "this is a {} car".format(self.engine)
```

```python
car1=Car(4,4,"petrol")
```

```python
car1.self_driving()
```
