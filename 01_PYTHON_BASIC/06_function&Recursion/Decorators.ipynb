# decorators

```python
#intial  decorators and clouser
def main_welcome(func):
    msg="hi vaibhav"
    def sub_function():
        print("hello there")
        func("hello vaibhav")
    return sub_function()
```

```python
main_welcome(print)
```

# decorators
passing function as parameter

```python
def main_welcome(func):
    msg="hi vaibhav"
    def sub_function():
        print("hello there")
        func()
    return sub_function()
```

```python
def student():
    print(" vaibhav is datascience student ")
```

```python
main_welcome(student)
```

```python
@main_welcome
def student():
    print(" vaibhav is datascience student ")
```
