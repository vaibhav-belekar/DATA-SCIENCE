# Filter function in python
The filter() function is used to select elements from a list based on a condition.
Key Points

Returns a filter object → convert using list()

Keeps only elements where condition is True

Often used with lambda

```python
def even(num):
    if num%2==0:
        return True
```

```python
lst=[1,2,3,4,5,6,7,8,9]
```

```python
filter(even,lst)
```

```python
list(filter(even,lst))
```

```python
filter(lambda num:num%2==0,lst)# using lambda function
```

```python
list(filter(lambda num:num%2==0,lst))
```
