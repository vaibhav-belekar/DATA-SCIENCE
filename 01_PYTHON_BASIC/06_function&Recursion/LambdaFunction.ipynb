# Python Lambda function
A Lambda function is a small, anonymous (no-name) function written in one line.
Key Points
Key Points

No function name

One line only

No return keyword

Used for short operations

```python
## normal function
def addition(num1,num2):
    return num1+num2
```

```python
addition(10,20)
```

```python
# lambda function
addition=lambda num1,num2:num1+num2
```

```python
addition(10,20)
```

```python
#normal function
def even(num):
    if num%2==0:
        return "even"
```

```python
#lambda function
even=lambda num:num%2==0
```

```python
even(25)
```
