# function in  python 
A function is a block of code that runs when you call it.
It helps you reuse code and keep your program clean.

```python
num=24
if num%2==0:
    print("the number is even ")
else:
    print("the number is odd")    
```

```python
def even_odd(num):
    if num%2==0:
        print("the number is even")
    else:
        print("the number is odd")    
        
        
```

```python
even_odd(55)
```

```python
def addition(num1,num2):
    return num1+num2
```

```python
val=addition(5,6)
```

```python
val
```

```python
# default parameter
def intro(name,age=18):
    print(f"my name is {name} and my age is {age}")
```

```python
intro("vaibhav",23)
```

```python
# positional argument and keyword argument
def hello(*args,**kwargs): 
    print(args)
    print(kwargs)
```

```python
hello("vaibhav","belekar",age=23,city="pune")
```

```python
list=['vaibhav','belekar']
dict={'age':23,'city':'pune'}
```

```python
hello(list,dict)
```

```python
hello(*list,**dict)
```

```python
list=[1,2,3,4,5,6,7]
```

```python
def evenoddsum(list):
    even_sum=0
    odd_sum=0
    for i in list:
        if i%2==0:
            even_sum+=i
        else:
            odd_sum+=i
    return even_sum,odd_sum            
```

```python
evenoddsum(list)
```
