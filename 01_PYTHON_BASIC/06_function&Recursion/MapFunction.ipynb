# map function
The map() function is used to apply a function to every element of a list (or iterable).

```python
def even_or_odd(num):
    if num%2==0:
        return "The number {} is even".format(num)
    else:
        return "The number {} is odd".format(num)
    
    
```

```python
even_or_odd(25)
```

```python
lst=[1,2,3,4,5,6,7,8,9]
```

```python
 list(map(even_or_odd,lst))
```
