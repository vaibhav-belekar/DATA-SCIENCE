## 1- Missing value


```python
# here we will  cheak the percentage of nan vallue present in each feature 
## 1 step make the list of feature which has missing values 
feature_with_na=[features for features in dataset.columns if dataset[features].isnull().sum()>1]

## 2 stepp print the feature name and the percentage of missing values
for feature in feature_with_na:
    print(feature,np.round(dataset[feature].isnull().mean(),4),'% missing value')
```

### since there are many missing values ,we need to find the relationship between missing values and Sales price
 
 -lets plots diagram for this relatindhip

```python
for feature in feature_with_na:
    data=dataset.copy()
    
    #lets make a variable that indicate 1 if the observation was missing or zero if observaation was present
    data[feature]=np.where(data[feature].isnull(),1,0)
    
    #lets calculaaate the mean saleprice where the information id m=nidding or present
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.show()
```

here with the relationship between the missing vallueas and the dependent variable is clarly visible , so we need to replace these nan values with someathing meaning which we will do in the feature engineering section

## numerical variable

```python
#list of numrical variables
numerical_feature=[feature for feature in dataset.columns if dataset[feature].dtype!='O']
print('numbers of numerical variables :',len(numerical_feature))
```

```python
dataset[numerical_feature].head()
```

# temporal variable (eg DateTime variable)


```python
#list of tempjoral variable that contain the year information
year_feature=[feature for feature in numerical_feature if 'Yr' in feature or 'Year' in feature]
year_feature
```

```python
#lets explore that contain year information
for feature in year_feature:
    print(feature,dataset[feature].unique())
```

```python
#lets analyse the Temporal datetime variable 
## we will cheakes wheter there is a relation between year and the hoouse is sold and

dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year sold')
plt.ylabel('median House price')
plt.title(" house Price vs YaerSold")
```

```python
# here we will compare the  diffrence between all years feature with year sold
for feature in year_feature:
    if feature!='YrSold':
        data=dataset.copy()
        #we will capture the diffrendce between year variable amd the house sold
        data[feature]=data['YrSold']-data[feature]
        
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

## Year Feature Analysis

- Converted year features into age using: YrSold - feature
- Reason: Age is more meaningful than raw year
- Plotted scatter plot between age and SalePrice
- Observed that newer houses tend to have higher prices

```python
#numerical variable are usally two types
## 1. continuas variable amd descrete variable
discrete_feature=[feature for feature in numerical_feature if len(dataset[feature].unique())<25 and feature not in year_feature+['Id']]
print(" Discrete Variable Count:{}".format(len(discrete_feature)))
```

```python
discrete_feature
```

```python
dataset[discrete_feature].head()
```

## Discrete Feature Selection

- Selected features with less than 25 unique values
- Reason: Such features behave like count or categorical variables
- Note: Threshold is a heuristic, not a strict rule

```python
## lets find the relationdhip between them and sale price
for feature in discrete_feature:
    data=dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

# continuos variable

```python
continuous_feature=[feature for feature in numerical_feature if feature not in discrete_feature+year_feature+['Id']]
print("continus feature {}".format(len(continuous_feature)))
```

```python
# lets analyse the cintinuos value by creating histogram to understand the 
for feature in continuous_feature:
    data=dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel("count")
    plt.title(feature)
    plt.show()
```
