```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head(2)
```

```python
## to cheak datatypes
final_df.dtypes
```

```python
country_name=final_df.Country.value_counts().index
country_name
```

```python
country_values=final_df.Country.value_counts().values
country_values
```

```python
#piechart -- top 3 country
plt.pie(country_values[:3],labels=country_name[:3],autopct='%1.2f%%')
```

Observation: zomato maximum record or transaction are from india
after thaat usa and then united kingdoms 

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:"Rating Counts"})
ratings
```

## observation :- 
1. when the rating  is between 4.5 to 4.9 --->excellent 
2. when the rating  is between 4.0 to 4.4 --->very good 
3. when the rating  is between 3.5 to 3.9 --->good 
4. when the rating  is between 3.0 to 3.4--->avg
5. when the rating  is between 2.5 to 2.9 --->avg
6. when the rating  is between 2.0 to 2.4 --->excellent 

 

```python
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x="Aggregate rating",y="Rating Counts",data=ratings,hue="Rating color")
```

```python
# plot exact same color 
sns.barplot(x="Aggregate rating",y="Rating Counts",data=ratings,hue="Rating color",palette=['black','red','orange','yellow','green','green'])
```

OBSERVATION:
1. not rated count is very high
2. maximum no if rating are between 2.5 to 3.4

```python
## count plot
sns.countplot(x="Rating color",data=ratings,palette=['black','red','orange','yellow','green','green'])
```
