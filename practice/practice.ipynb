```python
import pandas as pd

data = {
    'Restaurant': ['A','B','C','D','E','F','G','H'],
    'Country': ['India','India','USA','USA','India','UK','UK','USA'],
    'City': ['Mumbai','Delhi','NY','LA','Pune','London','Manchester','Chicago'],
    'Rating': [4.5, 3.0, 4.0, 2.5, 4.2, 3.8, 2.9, 4.1],
    'Rating color': ['Green','Yellow','Green','Red','Green','Yellow','Red','Green'],
    'Currency': ['INR','INR','Dollar','Dollar','INR','Pound','Pound','Dollar']
}

df = pd.DataFrame(data)
df
```

```python
# how many restorant in india
df[df['Country']=="India"].shape[0]
```

```python
# which restorant has green rating
df[df['Rating color']=='Green']["Restaurant"]
```

```python
# country wise restaurants
df.groupby('Country')['Restaurant'].count()
```

```python
# every contry avg rating 
df.groupby('Country')['Rating'].mean()
```

```python
#only green rating country name
df[df['Rating color']=='Green']['Restaurant']
```

```python
# in eevery country how many red rating restaurantds
df[df['Rating color']=="Red"].groupby('Country').size().reset_index(name='count')
```

```python
#whichb country has  more restarnts
df.groupby('Country').size().sort_values(ascending=False).head(1)
```

```python
df.groupby('Country')['City'].nunique()
```

```python
df.groupby('Country')['Rating'].max()
```

```python
df.groupby('Country')['Currency'].nunique()
```
