# to find the outlires


```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#define our dataset
data = [10,12,12,13,12,11,12,14,13,16,19,20,21,22,23,24,100,101,103,20,10,25]
```

```python
plt.hist(data)
```

```python
#z squre method
outliers=[]
def detect_outliirs(data):
    threshold=1
    mean=np.mean(data)
    std=np.std(data)
    for i in data:
        z_score=(i-mean)/std
        if np.abs(z_score)>threshold:
            outliers.append(i)
    return outliers
```

```python
detect_outliirs(data)
```

```python
# IQR method
dataset=sorted(data)
```

```python
q1,q3=np.percentile(dataset,[25,75])
```

```python
print(q1,q3)
```

```python
iqr=q3-q1
```

```python
print(iqr)
```

```python
# find lower and upper bound
lower_bound=q1-1.5*iqr
upper_bound=q3+1.5*iqr
```

```python
print(lower_bound,upper_bound)
```

```python
for i in dataset:
    if i<lower_bound or i>upper_bound:
        print(i)
```
