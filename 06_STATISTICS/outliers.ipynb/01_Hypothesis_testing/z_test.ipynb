# z test

Q1---> Suppose the IQ in a certain population is normally distributed with a mean of μ = 100 and standard deviation of σ = 15.

A researcher wants to know if a new drug affects IQ levels, so he recruits 20 patients to try it and records their IQ levels.

The following data represents the IQ levels of the 20 patients:

Data:
88, 92, 94, 94, 96, 97, 97, 97, 99, 99,
105, 109, 109, 109, 110, 112, 112, 113, 114, 115

```python
from statsmodels.stats.weightstats import ztest as ztest
# enter IQ levels for 20 patient
data=[88, 92, 94, 94, 96, 97, 97, 97, 99, 99,
105, 109, 109, 109, 110, 112, 112, 113, 114, 115]
```

```python
ztest(data,value=100)
```

```python
# this is p value np.float64(0.11012667014384257))
# if the p value<sigmificance 0.05 value reject null hypothesis  
```
