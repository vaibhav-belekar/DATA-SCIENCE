# join plot
A joint plot is a visualization that displays the relationship between two variables along with their individual distributions using scatter plots and histograms

### univariate analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind="hex")
```

```python
sns.jointplot(x="tip",y="total_bill",data=df,kind="reg")
```

# pairplot
A pairplot is used to visualize relationships between multiple variables in a dataset at once.

It creates a grid of plots showing:

-Scatter plots between every pair of features

-Distribution (histogram/KDE) on the diagonal

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue="sex")
```

# dist plot
dist plot is used to visualize the distribution of a single variable (how values are spread).

It shows:

Histogram (bars)

KDE curve (smooth density line)

```python
sns.distplot(df["tip"])
```

```python
sns.distplot(df["total_bill"],kde=False)
```

# categorical Plots
Categorical plots in Seaborn are used to visualize relationships between categorical variables and numerical data using plots like *barplot*, *countplot*, *boxplot*, and *violinplot*.

## -Countplot
A count plot displays how many times each category appears in the data.

```python
## countplot
sns.countplot(x='sex',data=df)
```

```python
sns.countplot(x='smoker',data=df,hue="sex")
```

## - Bar Plot
A bar plot shows the relationship between a categorical variable and a numerical variable using average values.


```python
sns.barplot(x='total_bill',y="sex",data=df)
```

## - BoxPlot
A box plot displays the spread of data using:

-Minimum

-Q1 (25%)

-Median (50%)

-Q3 (75%)

-Maximum

-Outliers

```python
sns.boxplot(x='smoker',y="total_bill",data=df)
```

```python
sns.boxplot(x='day',y="total_bill",data=df,hue="smoker",palette="rainbow")
```

## -Violin Plot
A violin plot is used to show the distribution of data along with its density (shape).

It is like a combination of box plot + KDE (density plot).

```python
sns.violinplot(x="total_bill",y="day",data=df,palette="rainbow")
```
