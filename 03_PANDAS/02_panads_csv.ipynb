```python
from io import StringIO, BytesIO
```

```python
data=('col1,col2,col3\n'
      'x,y,1\n'
      'a,b,2\n'
      'c,d,3\n')
```

```python
type(data)
```

```python
df=pd.read_csv(StringIO(data))
```

```python
df
```

```python
#read from specific columns
pd.read_csv(StringIO(data),usecols=['col1','col3'])
```
