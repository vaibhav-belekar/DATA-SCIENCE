# univariate analysis 

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('titanic_train.csv')
```

```python
df.head()
```

# 1. Categorical Data

## a. Countplot

```python
sns.countplot(x="Survived",data=df,hue="Sex")
```

```python
sns.countplot(x="Pclass",data=df,hue='Sex')
```

```python
sns.countplot(x='Survived',data=df,hue="Embarked")
```

## b. Pie chart

```python
df["Survived"].value_counts().plot(kind='pie',autopct='%1.2f%%')
```

```python
df["Pclass"].value_counts().plot(kind='pie',autopct='%1.2f%%')
```

# 2. Numerical data 

## a. Histogram

```python
plt.hist(df['Age'],bins=25)
```

```python
sns.histplot(df['Age'],bins=25)
```

## b. Distplot

```python
sns.distplot(df['Age'])
```

## c. BoxPlot

```python
sns.boxplot(df['Age'])
```
